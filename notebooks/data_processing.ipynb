## Data processing
### Bring all the None Values to the same type -> for LLM easier then!
### Lowercase everything!
### remoce white useless spaces

```python
df["locations"].unique()
```

```python
import pandas as pd

missing_values = ["", " ", "nan", "null", "na", "n/a", "none", "-", "--"]

def clean_value(x):
    # 1. echte numpy.nan
    if isinstance(x, float) and pd.isna(x):
        return "n/a"

    # 2. echte Python None
    if x is None:
        return "n/a"

    # 3. Listen: rekursiv jeden Eintrag bereinigen
    if isinstance(x, list):
        return [clean_value(i) for i in x]

    # 4. Dictionaries: rekursiv Einträge bereinigen
    if isinstance(x, dict):
        return {k: clean_value(v) for k, v in x.items()}

    # 5. Strings normalisieren
    x = str(x).strip().lower()

    # 6. Strings, die wie Missing aussehen → "n/a"
    if x in missing_values:
        return "n/a"

    return x

# Apply to entire dataframe
for col in df.columns:
    df[col] = df[col].apply(clean_value)
```

# We need to build a a text line for the LLM -> we do that now

```python
important_cols = [
    "career_objective",
    "skills",
    "professional_company_names",
    "positions",
    "responsibilities",
    "responsibilities.1",
    "educational_institution_name",
    "degree_names",
    "major_field_of_studies",
    "passing_years",
    "extra_curricular_activity_types",
    "extra_curricular_organization_names",
    "extra_curricular_organization_links",
    "languages",
    "proficiency_levels",
    "certification_skills",
    "certification_providers",
    "matched_score"
]

df[important_cols] = df[important_cols].fillna("")

def build_cv_text(row):

    # Define what counts as missing
    missing_tokens = {"", "none", "n/a", "nan", None}

    def clean_value(v):
        if v is None:
            return None

        # Lists: clean recursively
        if isinstance(v, list):
            cleaned = [clean_value(x) for x in v]
            cleaned = [x for x in cleaned if x not in missing_tokens and x is not None]
            return cleaned if cleaned else None

        # Convert to string
        v = str(v).strip().lower()
        if v in missing_tokens:
            return None

        return v

    parts = ["### CV ENTRY"]

    for col in important_cols:
        header = col.replace("_", " ").title()
        val = clean_value(row[col])

        # CASE 1: Missing section → generate fallback
        if val is None:
            parts.append(f"\n### {header}:\n(no information provided)")
            continue

        # CASE 2: Lists → bullet points
        if isinstance(val, list):
            bullet_list = "\n".join([f"- {item}" for item in val])
            parts.append(f"\n### {header}:\n{bullet_list}")
            continue

        # CASE 3: Normal string values
        parts.append(f"\n### {header}:\n{val}")

    return "\n".join(parts)

df["cv_text"] = df.apply(build_cv_text, axis=1)

output_path = "../data/cv_texts.txt"

with open(output_path, "w", encoding="utf-8") as f:
    for i, text in enumerate(df["cv_text"], start=1):
        f.write(f"CV {i}: {text}\n")
```
